```python
import re

month_map = {
    "january": "01", "jan": "01",
    "february": "02", "feb": "02",
    "march": "03", "mar": "03",
    "april": "04", "apr": "04",
    "may": "05",
    "june": "06", "jun": "06",
    "july": "07", "jul": "07",
    "august": "08", "aug": "08",
    "september": "09", "sep": "09", "sept": "09",
    "october": "10", "oct": "10",
    "november": "11", "nov": "11",
    "december": "12", "dec": "12"
}
```

```python
# Function to parse the date from the input text
def parse_date(text):
    # match the date formats using regex
    date_patterns = [
        r'(\d{1,2})(?:st|nd|rd|th)?[ \-/](\w+)[ \-/](\d{4})',  # Ex- 5th March 2023, 5/Mar/2023, 5-March-2023
        r'(\w+)[ \-/](\d{1,2})(?:st|nd|rd|th)?,?[ \-/](\d{4})',  # Ex- March 5, 2023, Mar-5-2023
        r'(\d{4})[ \-/](\d{2})[ \-/](\d{2})',  # Ex- 2023-03-05
        r'(\d{2})[ \-/](\d{2})[ \-/](\d{4})',  # Ex- 05/03/2023
        r'(\d{4})[ \-/](\d{1,2})[ \-/](\d{1,2})',  # Ex- 2023/3/5
        r'(\d{1,2})(?:st|nd|rd|th)? of (\w+),? (\d{4})', # Ex- 1st of January 2021
        r'(\d{1,2})(?:st|nd|rd|th)?[ \-](\w+),? (\d{4})', # Ex- 15th January, 2021
        r'(\d{1,2})[./-](\d{1,2})[./-](\d{2,4})', # Ex- 02.01.2021
        r'(\d{4})[./-](\d{1,2})[./-](\d{1,2})', # Ex- 2021.01.10
        r'(\d{1,2})[./-](\d{1,2})[./-](\d{2,4})' # Ex- 12/12/12
    ]

    for pattern in date_patterns:
        match = re.search(pattern, text)
        if match:
            groups = match.groups()
            if len(groups) == 3:
                if groups[0].isalpha():  # Format: Month Day, Year
                    day = groups[1].zfill(2)
                    month = month_map.get(groups[0].lower(), groups[0])
                    year = groups[2]
                elif groups[1].isalpha():  # Format: Day Month Year
                    day = groups[0].zfill(2)
                    month = month_map.get(groups[1].lower(), groups[1])
                    year = groups[2]
                elif len(groups[0]) == 4:  # Format: Year Month Day
                    year = groups[0]
                    month = groups[1].zfill(2)
                    day = groups[2].zfill(2)
                else:  # Format: Day Month Year
                    day = groups[0].zfill(2)
                    month = groups[1].zfill(2)
                    year = groups[2]

                # Return the formatted date
                return f"{day}/{month}/{year}"

    return None
```

```python
df['Parsed Output'] = df['Input'].apply(parse_date)
```
